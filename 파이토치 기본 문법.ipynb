# 텐서 수학 및 비교 연산
---

```python
x = torch.tensor([1,2,3])
y = torch.tensor([9,8,7])
```

```python
# 덧셈
z1 = torch.empty(3)
torch.add(x, y, out = z1)
print(z1,'\n')

z2 = torch.add(x,y)
print(z2,'\n')

z = x + y 
print(z,'\n')
```

```python
# 뺄셈
z = x - y
print(z,'\n')
```

```python
# 나누기
z = torch.true_divide(x,y)
print(z,'\n')
```

```python
# inplace operations
t = torch.zeros(3)
t.add_(x)
# t += x
print(t,'\n')
```

```python
# 지수화
z = x.pow(2)
# z = x **2
print(z)
```

```python
z = x > 0
print(z)

z = x < 0
print(z)
```

```python
# 행렬의 곱셈
x1 = torch.rand((2,5))
x2 = torch.rand((5,3))

x3 = torch.mm(x1,x2) 
# x3 = x1.mm(x2)
print(x3.shape)
```

```python
# 행렬 지수화
matrix_exp = torch.rand(5,5)
print(matrix_exp.matrix_power(3))
```

```python
z = x * y
print(z,'\n')

z = torch.dot(x,y)
print(z,'\n')
```

```python
# 배치 행렬 
batch = 32 
n = 10 
m = 20
p = 30

tensor1 = torch.rand((batch, n, m)) # 3차원 텐서
print(tensor1.shape, '\n')

tensor2 = torch.rand((batch, m, p))
print(tensor2.shape, '\n')

# 행렬 곱셈
out_bmm = torch.bmm(tensor1, tensor2) # (batch, n, p)
print(out_bmm.shape, '\n')
```

# 유용한 텐서 연산
---

```python
# 합산해야 하는 차원을 지정
print(x, '\n')

# 합
sum_x = torch.sum(x, dim = 0)
print(sum_x, '\n')

# 가장 큰 값과 인덱스
values, index = torch.max(x, dim = 0)
print(values, index, '\n')

# 가장 작은 값과 인덱스
values, index = torch.min(x, dim = 0)
print(values, index, '\n')

# 절대값
abs_x = torch.abs(x)
print(abs_x,'\n')
```

```python
# 최대값의 인덱스
z = torch.argmax(x, dim=0)
print(z,'\n')

# 최소값의 인덱스
z = torch.argmin(x, dim=0)
print(z,'\n')

# 평균
mean_x = torch.mean(x.float(), dim = 0)
print(mean_x,'\n')

# equal
z = torch.eq(x,y)
print(z)
```

```python
sort, index = torch.sort(y, dim=0, descending = True)
print(sort, index)
```

```python
# 0 보다 큰 x의 요소를 확인하고 최소값으로 변경
z = torch.clamp(x, min = 0)
print(z,'\n')

z = torch.clamp(x, min = 2)
print(z, '\n')

z = torch.clamp(x, max = 1)
print(z, '\n')
```

```python
x = torch.tensor([1,3,1,0,1], dtype = torch.bool)
print(x,'\n')

# 하나라도 True인가?
z = torch.any(x)
print(z,'\n')

# 모두 통일 되는가?
z = torch.all(x)
print(z,'\n')
```

# 텐서의 인덱싱
---

```python
batch_size = 10
features = 25

x = torch.rand((batch_size, features))
print(x.shape, '\n')
print(x[0].shape, '\n') # x[0,:]
print(x[:,0].shape, '\n')
print(x[2,0:15].shape, '\n')
```

```python
x = torch.arange(10,20)
index  = [2,5,7]

print(x[index])
```

```python
x = torch.rand((3,5))
rows = torch.tensor([1,2])
cols = torch.tensor([4,1])
print(x,'\n')
print(x[rows,cols]) # 2행 5열 / 3행 2열
```

```python
x = torch.arange(10)
print(x[(x < 2) | (x > 8)])

# 짝수 요소
print(x[x.remainder(2) == 0 ])

# 홀수 요소
print(x[x.remainder(2) == 1 ])
```

```python
# x가 5보다 큰 요소를 먼저 찾고 
# 5보다 크다면 그대로
# 5보다 작다면 곱하기2
print(torch.where(x > 5, x, x*2))
```

```python
# 중복없는 고유의 값만 반환
print(torch.tensor([0,0,1,2,2,3,4,3,5]).unique())
```

```python
# 몇 차원인가?
x.ndimension()
```

```python
# 요소의 수
x.numel()
```

```python
a = torch.tensor([[1,2],[2,4],[5,1]])
print(a.ndimension())
print(a.numel())
```

# reshape
----

```python
x = torch.arange(9)
print(x,'\n')

x_33 = x.view(3,3)
print(x_33,'\n')

y = x_33.t()
print(y,'\n')

print(y.contiguous().view(9),'\n')
```

```python
x1 = torch.rand((2,5))
x2 = torch.rand((2,5))

print(torch.cat((x1,x2), dim=0).shape,'\n')
print(torch.cat((x1,x2), dim=1).shape,'\n')
```

```python
z = x1.view(-1)
print(z.shape)
```

```python
batch = 64
x = torch.rand((batch, 2, 5, 5))
z = x.view(batch,-1)

print(x.shape,'\n')
print(z.shape,'\n')
```

```python
z = x.permute(0,2,1,3)
print(z.shape)
```

```python
x = torch.arange(10)

print(x.unsqueeze(0).shape,'\n')
print(x.unsqueeze(1).shape,'\n')
```

```python
x = torch.arange(10).unsqueeze(0).unsqueeze(1)
print(x.shape,'\n')

z = x.squeeze(1)
print(z.shape,'\n')
```
